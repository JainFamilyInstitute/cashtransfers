```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import warnings
from skimage import io, color, transform, feature
%matplotlib inline
```

```python
df = pd.read_csv('/Volumes/Samsung_t3/project-repos/jfi/studentloans/data/data.csv')
```

```python
# rename columns
df.columns = ['paper', 'summ', 'url', 'country', 'allocation', 'NOCT', 'involved', 'tools', 'relevance', 'economic', 'psych', 'social', 'health', 'schooling', 'means', 'research_design', 'employment_effects']
```

```python
df = df.fillna("")
df['size'] = 40000
```

```python
cols = ['economic', 'psych', 'social', 'health', 'schooling']
df[cols] = df[cols].replace({'Y':1, "":0})
```

```python
# Original
df.head(3)
```

```python
#or set columns names by list
s = df[cols].stack()
df1 = pd.get_dummies(s[s == 1].reset_index(level=1).drop(0, axis=1), prefix='', prefix_sep='')
```

```python
df2 = pd.merge(df, df1, on=cols, left_index=True, right_index=True)
```

```python
# Duplicate entries with more than 1 tag; break out tags
df2
```

```python
df2.shape
```

```python
df3 = df[df[cols].sum(axis=1) == 1]
```

```python
# Get all rows that required no duplicating
df3.head(5)
```

```python
# Join subsets
df4 = pd.concat([df2,df3])
df4 = df4.fillna("")
```

```python
# Reorder columns
col_order = ['paper', 'summ', 'url', 'country', 'allocation', 'NOCT', 'involved', 'tools', 'relevance', 'economic', 'psych', 'social', 'health', 'schooling', 'means', 'research_design', 'employment_effects', 'size']
df4 = df4[col_order]
```

```python
df4.loc[df4['economic']==1, "path"] = "https://raw.githubusercontent.com/3milychu/jfi/master/studentloans/assets/1.jpg"
df4.loc[df4['psych']==1, "path"] = "https://raw.githubusercontent.com/3milychu/jfi/master/studentloans/assets/2.jpg"
df4.loc[df4['social']==1, "path"] = "https://raw.githubusercontent.com/3milychu/jfi/master/studentloans/assets/3.jpg"
df4.loc[df4['health']==1, "path"] = "https://raw.githubusercontent.com/3milychu/jfi/master/studentloans/assets/4.jpg"
df4.loc[df4['schooling']==1, "path"] = "https://raw.githubusercontent.com/3milychu/jfi/master/studentloans/assets/5.jpg"
```
